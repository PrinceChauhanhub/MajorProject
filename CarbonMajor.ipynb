# Feature Scaling / Normalization

```python
import ast

def create_dummy_variables(X, column_name):
    X = X.copy()

    unique_values = set()


    for item in X[column_name].dropna():
        values = ast.literal_eval(item)
        unique_values.update(values)


    for value in unique_values:
        X[value] = X[column_name].apply(lambda x: int(value in ast.literal_eval(x)) if pd.notnull(x) else 0)


    X.drop(columns=column_name, inplace=True)

    return X, unique_values
```

```python
import pickle

X, recycling_categories = create_dummy_variables(X, 'recycling')
X, cooking_categories = create_dummy_variables(X, 'cooking_with')

dummy_info = {
    "recycling": recycling_categories,
    "cooking_with": cooking_categories
}

with open("dummy_info.pkl", "wb") as file:
    pickle.dump(dummy_info, file)
```

```python
X.head(2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder


ordinal_cols = ['body_type', 'shower_frequency', 'social_activity', 'air_travel_frequency', 'waste_bag_size', 'energy_efficiency']
onehot_cols = ['gender', 'diet', 'heating_energy_source', 'transport']


preprocessor = ColumnTransformer(
    transformers=[
        ('ordinal', OrdinalEncoder(), ordinal_cols),
        ('onehot', OneHotEncoder(drop='first'), onehot_cols)
    ],
    remainder='passthrough'
)


X_transformed = preprocessor.fit_transform(X)

```

```python
# Convert back to DataFrame with proper column names
ohe_feature_names = preprocessor.named_transformers_['onehot'].get_feature_names_out(onehot_cols)
all_feature_names = ordinal_cols + list(ohe_feature_names) + [col for col in X.columns if col not in ordinal_cols + onehot_cols]
X_df = pd.DataFrame(X_transformed, columns=all_feature_names)
```

```python
X_df.reset_index(drop=True, inplace=True)
```

```python
with open('preprocessor.pkl', 'wb') as f:
    pickle.dump(preprocessor, f)
```

```python

selected_features = ordinal_cols + onehot_cols + [col for col in df_cleaned.columns if col not in ordinal_cols + onehot_cols]
with open('selected_features.pkl', 'wb') as f:
    pickle.dump(selected_features, f)
```

# train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_df, y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape, y_train.shape
```

```python
X_train.columns
```

# Model Training

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'SVR': SVR(),
    'KNN': KNeighborsRegressor(),
    'Ridge': Ridge(),
    'Lasso': Lasso()
}

best_model_name = None
best_model = None
best_r2_score = -np.inf

# Train & Evaluate
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"{name}:")
    r2 = r2_score(y_test, y_pred)
    print(f"  R Score: {r2:.4f}")

    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    print(f"  RMSE: {rmse:.4f}")

    mean_y = np.mean(y_train)
    relative_rmse = (rmse / mean_y) * 100
    print(f"  Relative RMSE: {relative_rmse:.4f}")
    print("-" * 30)

    if r2 > best_r2_score:
        best_r2_score = r2
        best_model_name = name
        best_model = model

print(f"Best model is: {best_model_name}. Generating residual plot...")

y_pred_best = best_model.predict(X_test)
residuals_best = y_test - y_pred_best


sns.residplot(x=y_pred_best, y=residuals_best, lowess=True, line_kws={'color': 'red'})
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title(f'Residual Plot for {best_model_name}')
plt.show()
```
